```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial as fat
```

```python
def comb(n, i):
    if i >= 0 and i <= n:
        return fat(n) / (fat(i) * fat(n - i))
    
def B(n, i, t):
    return comb(n, i) * t**i * (1 - t)**(n-i)

def getM(n):
    pass
```

```python
def plotBezierCurve(c, step=0.01):
    n = len(c)
    pontos = []
    for t in np.arange(0, 1 + step, step):
        p = np.zeros(2)
        for i in range(n):
            p += c[i] * B(n, i+1, t)
        pontos.append(p)
    
    pontos = np.array(pontos)
    
    plt.plot(c[:, 0], c[:, 1], 'b--', label='Pontos de Controle')
    plt.plot(pontos[:, 0], pontos[:, 1], 'r-', label='Curva de BÃ©zier')
    plt.legend()
    plt.show()
```

```python
p = np.array([
    (0, 0),
    (0.25, 1),
    (0.5, 1.5),
    (0.75, 1),
    (1, 0)
])

plotBezierCurve(p, step=0.01)
```
